```python
sns.set(style="whitegrid")

# List of categorical features to analyze
categorical_features = ['Sex', 'ChestPainType', 'ExerciseAngina', 'ST_Slope', 'RestingECG']

# Set up the plot grid
fig, axs = plt.subplots(3, 2, figsize=(14, 12))
axs = axs.flatten()

# Plot count plots for each feature
for i, feature in enumerate(categorical_features):
    sns.countplot(data=df, x=feature, hue='HeartDisease', palette='Set2', ax=axs[i])
    axs[i].set_title(f'{feature} vs HeartDisease')
    axs[i].legend(title='HeartDisease')

# Remove the empty subplot if needed
fig.delaxes(axs[-1])  # Only if there's an unused subplot

plt.tight_layout()
plt.show()
```
